```python
import pandas as pd
df = pd.read_csv('WA_Fn-UseC_-HR-Employee-Attrition.csv')
df.head()
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(2, 2, figsize=(18, 12))

# -----1. Attrition Count-----
attrition = df['Attrition'].value_counts()
axes[0,0].bar(attrition.index, attrition.values)
axes[0,0].set_title('employees Attrition')

# -----2. Attrition by Department----
dept = pd.crosstab(df['Department'], df['Attrition'])
dept.plot(kind='bar', stacked=True, ax=axes[0,1])
axes[0,1].set_title('Attrition by Department')

# -----3. Age Distribution------
axes[1,0].hist(df['Age'], bins=20)
axes[1,0].set_title('Age Distribution')

# ----4. Income vs Attrition (better than raw plot)--
income = df.groupby('Attrition')['MonthlyIncome'].mean()
axes[1,1].bar(income.index, income.values)
axes[1,1].set_title('Income vs Attrition')

plt.tight_layout()
plt.savefig('hr-dashboard.png',dpi=300, bbox_inches='tight')
plt.show()

print('Total Employees:', len(df))
print('Employees left:', len(df[df['Attrition']=='Yes']))
print('Employees stayed:', len(df[df['Attrition']=='No']))


```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(2, 2, figsize=(18, 12))

# <> 1. Attrition %
attrition = df['Attrition'].value_counts(normalize=True) * 100
axes[0,0].pie(attrition.values, labels=attrition.index, autopct='%1.1f%%')
axes[0,0].set_title('Attrition %')

# <> 2. Attrition by Department
dept = pd.crosstab(df['Department'], df['Attrition'])
dept.plot(kind='bar', stacked=True, ax=axes[0,1])
axes[0,1].set_title('Attrition by Department')

# <> 3. Age vs Attrition
df[df['Attrition']=='Yes']['Age'].plot(kind='hist', alpha=0.7, ax=axes[1,0])
df[df['Attrition']=='No']['Age'].plot(kind='hist', alpha=0.5, ax=axes[1,0])
axes[1,0].set_title('Age Distribution by Atrrition')
axes[1,0].legend('Left', 'Stayed')

# <> 4. Job Role vs Attrition (Top roles)
role = pd.crosstab(df['JobRole'], df['Attrition'])
role.sort_values(by='Yes', ascending=False).head(5).plot(kind='bar', ax=axes[1,1])
axes[1,1].set_title('Top Job Roles with Attrition')

total = len(df)
left = len(df[df['Attrition']=='Yes'])
stay = len(df[df['Attrition']=='No'])

print(f'Total Employees: {total}')
print(f'Attrition Rate: {round(left/total*100, 2)}%')
print(f'Employees left: {left}')
print(f'Employees stayed: {stay}')

plt.suptitle('HR Analytics Dashboard', fontsize=20)
plt.tight_layout()
plt.savefig('hr-dashboard_pro.png', dpi=300, bbox_inches='tight')
plt.show()
```
